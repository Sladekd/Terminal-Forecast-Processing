```python
# This notebook helps users to clean Ogimet files into pandas with TAFs
import pandas as pd
import re

def clean_taf_ogimet_file(file_path):
  cleaned_tafs = []

  with open(file_path, 'r') as f:
    text = f.read()

    # Use regular expressions to find all TAF forecasts
    # assume that there is YYYYMMDDHHMM pattern
    # - assume that  TAF is terminated by =
    taf_forecasts = re.findall(r'\d{12}.+?=', text, re.DOTALL)

    # Remove the junk data in between TAF forecasts and add to the list
    clean_taf_forecasts = [taf.strip() for taf in taf_forecasts]
    cleaned_tafs.extend(clean_taf_forecasts)

  # Create a DataFrame from the cleaned TAF forecasts list
  df = pd.DataFrame(cleaned_tafs, columns=['TAF Forecast'])

  return df
```
